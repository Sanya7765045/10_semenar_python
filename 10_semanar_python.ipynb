```python
sys.scatterplot(data=penguins, x="bill_length_mm", y='bill_depth_mm', hue="island", size = 20)
```

```python
from seaborn.axisgrid import PairGrid
peng = ['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']
p = sys.PairGrid(penguins[peng])
p.map(sys.scatterplot)
```

```python
sys.histplot(data=penguins, x="body_mass_g", hue = 'bill_length_mm')
```

```python
sys.histplot(data=penguins, x="island", hue = 'sex')
```

```python
penguins.loc[penguins['bill_length_mm'] >= 42, 'bill_group'] = 'high'
penguins.loc[(penguins['bill_length_mm'] >= 35) & (penguins['bill_length_mm'] < 42), 'bill_group'] = 'middle'
penguins.loc[penguins['bill_length_mm'] < 35, 'bill_group'] = 'low'
```

```python
penguins.groupby('bill_group')['bill_length_mm'].mean().plot(kind = 'bar')
```

```python
sys.displot(data = penguins, x = 'flipper_length_mm',hue = 'bill_group')
```

В ячейке ниже представлен код генерирующий DataFrame, которая состоит всего из 1 столбца. Ваша задача перевести его в one hot вид. Сможете ли вы это сделать без get_dummies?





```python
import random
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI': lst})
data.head()
```

```python
data.loc[data['whoAmI'] == 'human', 'human'] = '1'
data.loc[data['whoAmI'] == 'robot', 'human'] = '0'
data.loc[data['whoAmI'] == 'robot', 'robot'] = '1'
data.loc[data['whoAmI'] == 'human', 'robot'] = '0'
data[['human', 'robot']]
```

```python
pd.get_dummies(data['whoAmI'])
```
